### Task 7

Приєднайте три набори даних із завдань 1-6 до нового набору даних (використовуючи перетин назв країн).
- Використовуйте лише дані про ВВП за останні 10 років (2006-2015 рр.) і
лише 15 найкращих країн за рейтингом Scimagojr (Rank від 1 до 15)
- Індексом цього DataFrame повинна бути назва країни, а стовпцями мають бути ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
'2015'].
Ви повинні отримати DataFrame з 15 рядками та 20 стовпцями.

```python
# energy_df - дані про енергопостачання
# gdp_df - дані про ВВП
# scimagojr_df - дані про рейтинг Scimagojr

# Крок 1: Вибрати 15 країн з найвищим рейтингом Scimagojr
top_15_scimagojr = scimagojr_df[scimagojr_df['Rank'] <= 15]

# Крок 2: Обмежити дані про ВВП за останні 10 років (2006-2015)
gdp_last_10_years = csv_df[['Country', 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]]

# Крок 3: Приєднати дані про енергопостачання та Scimagojr на основі країни
energy_scimagojr_merge = pd.merge(top_15_scimagojr, df, how='inner', left_on='Country', right_on='Country')

# Крок 4: Приєднати дані про ВВП до вже злитих даних
final_df = pd.merge(energy_scimagojr_merge, gdp_last_10_years, how='inner', on='Country')

# Крок 5: Встановити країни як індекс
final_df.set_index('Country', inplace=True)

# Крок 6: Вибрати лише необхідні стовпці
final_columns = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
                 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
                 '% Renewable', 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]


final_df = final_df[final_columns]
print(final_df)
```

### Task 8

Створіть функцію, щоб визначити середній ВВП країн за останні 10 років.

Ця функція повинна повертати об’єкт Series з 15 країнами та їх середній ВВП, відсортований за спаданням.


```python
def calculate_average_gdp(df):
    """
    function calculate average gdp for each country for the last 10 years (2006-2015).
    
    Input:
    df (DataFrame): DataFrame, contain data about GDP with columns from 2006 till 2015.

    Return:
    Series: average gdp for each country, sorting by desc.
    """
    gdp_columns = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
    average_gdp = df[gdp_columns].mean(axis=1)
    
    average_gdp_series = pd.Series(average_gdp, index=df.index, name='Average GDP')
    sorted_average_gdp = average_gdp_series.sort_values(ascending=False)
    
    return sorted_average_gdp
```

```python
average_gdp= calculate_average_gdp(final_df)

print(average_gdp)
```

### Task 9

Створіть функцію, щоб визначити, наскільки змінився ВВП за останні 10 років для країни з 5-м середнім ВВП.


Ця функція повинна повертати кортеж із назвою країни та значенням, на яке змінився ВВП.

```python
def gdp_change_for_5th_country(df):
    """
    Function define, how GDP has changed for the last 10 years for country with 5-th evarage GDP.

    Input:
    df (DataFrame): DataFrame, which contains data about GDP with col from 2006 till 2015.

    Return:
    tuple: Country name and value changes GDP (2015 - 2006).
    """
    sorted_average_gdp = calculate_average_gdp(df)
    fifth_country = sorted_average_gdp.index[4]  
    
    gdp_2006 = df.loc[fifth_country, 2006]
    gdp_2015 = df.loc[fifth_country, 2015]
    
    gdp_change = gdp_2015 - gdp_2006
    
    return (fifth_country, gdp_change)
```

```python
country_name, change_in_gdp = gdp_change_for_5th_country(final_df)

print(f"Country: {country_name}, Change GDP: {change_in_gdp}")
```

### Task 10

Створіть функцію, щоб визначити, в якій країні встановлено максимум поновлюваних джерел енергії (% Renewable) та який саме відсоток.

Ця функція повинна повертати кортеж із назвою країни та відсотком.

```python
def country_with_max_renewable(df):
    """
    Function determine country with max % renovated enery sources (% Renewable).

    Input:
    df (DataFrame): DataFrame, which contains data about renovated enery sources (% Renewable).

    Return:
    tuple: Country name and % renovated enery sources.
    """
    max_renewable_country = df['% Renewable'].idxmax()  
    max_renewable_value = df.loc[max_renewable_country, '% Renewable']  
    
    return (max_renewable_country, max_renewable_value)
```

```python
country_name, renewable_percentage = country_with_max_renewable(final_df)

print(f"Country with the biggest % renovated energy sources : {country_name}, Percentage: {renewable_percentage}%")
```

### Task 11

Створіть стовпець, в якому оцінюється чисельність населення. 

Для пошуку оцінки використовуйте значення енергопостачання (Energy Supply) та енергопостачання на душу населення (Energy Supply per capita). Яка країна є шостою за цією оцінкою?

Ця функція повинна повертати кортеж із назвою країни та населенням

```python
def estimate_population(df):
    """
    Creates a column to estimate population based on
    values of energy supply (Energy Supply) and energy supply per capita
    (Energy Supply per Capita), and also determines the sixth most populous country.

    Arguments:
    df (DataFrame): DataFrame containing columns 'Energy Supply' and 'Energy Supply per Capita'.

    Returns:
    tuple: The name of the sixth largest country and its population.
    """
    df['Estimated Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
   
    df_sorted = df.sort_values(by='Estimated Population', ascending=False)
    sixth_country = df_sorted.iloc[5]  
    
    return (sixth_country.name, sixth_country['Estimated Population'])
```

```python
sixth_country, population = estimate_population(final_df)

print(f"Шоста країна за чисельністю населення: {sixth_country}, Населення: {population:.0f}")
```

### Task 12

Створіть стовпець, в якому оцінюється кількість цитованих документів на одну особу. 

Який взаємозв'язок між кількістю цитованих документів на душу населення та енергозабезпеченням на душу населення? Використовуйте
метод .corr () (кореляція Пірсона).

```python
def calculate_citations_per_person(final_df):
    """
    Creates a column to estimate the number of cited documents per person.
    Also calculates the correlation between the number of cited documents per capita
    and energy supply per capita.

    Input:
    df (DataFrame): DataFrame containing the columns 'Citations', 'Self-citations', 'Estimated Population', 'Energy Supply per Capita'.

    Return:
    float: Pearson correlation between the number of cited documents per capita
    and energy supply per capita.
    """
    
    final_df['Citable Documents per Person'] = final_df['Citable documents'] / final_df['Estimated Population']

    correlation_matrix = final_df[['Citable Documents per Person', 'Energy Supply per Capita']].corr(method='pearson')
    
    # correlation between'Citable Documents per Person' та 'Energy Supply per Capita'
    correlation_value = correlation_matrix.loc['Citable Documents per Person', 'Energy Supply per Capita']

    return correlation_value
```

```python
correlation_value = calculate_citations_per_person(final_df)

print(f"Кореляція між цитованими документами на душу населення та енергозабезпеченням на душу населення: {correlation_value}")
```

### Task 13

Створіть новий стовпчик в який занесіть 1, якщо значення % відновлюваної енергії (% Renewable) країни дорівнює або вище медіани для всіх країн, і 0, якщо значення (% Renewable) нижче медіани.


Ця функція повинна повертати Series, індексом якого є назва країни, відсортована у порядку зростання Rank

```python
final_df.columns
```

```python
final_df
```

```python
def renewable_energy_flag(df):
    """
    Creates a new column labeled where 1 is if the country's % renewable energy
    equal to or above the median, and 0 if below the median. Returns Series sorted by Rank.

    Arguments:
    df (DataFrame): DataFrame containing column '% Renewable' and 'Rank'.

    Returns:
    Series: Series with 1 and 0, the index is the name of the country, sorted by increasing Rank.
    """
     
    renewable_median = df['% Renewable'].median()
    df['Renewable Flag'] = (df['% Renewable'] >= renewable_median).astype(int)

    return df[['Renewable Flag']]
```

```python
renewable_series = renewable_energy_flag(final_df)

print(renewable_series)
```

### Task 14

Використайте наступний словник, щоб згрупувати країни за континентами, потім створіть DataFrame, який відображає розмір вибірки (кількість країн на кожному континенті), а також суму, середнє та стандартне відхилення для оцінки кількості населення для кожної країни.

ContinentDict = {'China':'Asia',

'United States':'North America',

'Japan':'Asia',

'United Kingdom':'Europe',

'Russian Federation':'Europe',

'Canada':'North America',

'Germany':'Europe',

'India':'Asia',

'France':'Europe',

'South Korea':'Asia',

'Italy':'Europe',

'Spain':'Europe',

'Iran':'Asia',

'Australia':'Australia',

'Brazil':'South America'}

Ця функція повинна повертати DataFrame з індексом, що має ім’я Continent ['Asia', 'Australia', 'Europe', 'North America', 'South America'] та стовпцями ['size', 'sum', 'mean', 'std '].

```python
ContinentDict = {
    'China':'Asia',
    'United States':'North America',
    'Japan':'Asia',
    'United Kingdom':'Europe',
    'Russian Federation':'Europe',
    'Canada':'North America',
    'Germany':'Europe',
    'India':'Asia',
    'France':'Europe',
    'South Korea':'Asia',
    'Italy':'Europe',
    'Spain':'Europe',
    'Iran':'Asia',
    'Australia':'Australia',
    'Brazil':'South America'
}
```

```python
def continent_population_stats(df):
    """
    Create DataFrame, shows size, sum, median and standard deviation of the population by continent.

    Input:
    df (DataFrame): DataFrame, which contains column 'Population Estimate' (population estimate).

    Return:
    DataFrame: DataFrame with index 'Continent' and columns['size', 'sum', 'mean', 'std'].
    """
    df['Continent'] = df.index.to_series().map(ContinentDict)

    continent_stats = df.groupby('Continent')['Estimated Population'].agg(['size', 'sum', 'mean', 'std'])
    
    return continent_stats
```

```python
continent_stats_df = continent_population_stats(final_df)

print(continent_stats_df)
```

### Task 15

Створіть бульбашкову діаграму, що відображає залежність % поновлюван их джерел від рейтингу країни (% Renewable vs. Rank). 

Розмір бульбашки від повідає ВВП країн 2015 року, а колір відповідає континенту.

```python
import seaborn as sns

required_columns = ['Rank', '% Renewable', 2015, 'Continent']

# style and context 
sns.set(style="whitegrid", context="talk")

plt.figure(figsize=(14, 10))

# Plotting the bubbles
bubble_plot = sns.scatterplot(
    x='Rank', 
    y='% Renewable', 
    size=2015,        
    hue='Continent',  
    data=final_df, 
    palette={'Asia': 'red', 'North America': 'blue', 'Europe': 'green', 
             'Australia': 'yellow', 'South America': 'orange'},
    sizes=(100, 2000), 
    alpha=0.6, 
    edgecolor="w", 
    linewidth=2
)

# Annotate countries with conditional display
for i in range(final_df.shape[0]):
    country_name = final_df.index[i]
    gdp_value = final_df[2015].iloc[i] 
    if pd.notnull(gdp_value):  
        # show bubble
        plt.text(final_df['Rank'].iloc[i], final_df['% Renewable'].iloc[i], 
                 final_df.index[i], fontsize=10, ha='center')
    else:  
        # GDP == null, country name without bubble
        plt.text(final_df['Rank'].iloc[i], final_df['% Renewable'].iloc[i], 
                 country_name, fontsize=10, ha='center')

# labels and title
plt.xlabel('Rank')
plt.ylabel('% Renewable')
plt.title('Bubble Chart')

plt.show()
```

```python
continent_colors = {
    'Asia': 'red',
    'Europe': 'green',
    'North America': 'blue',
    'South America': 'orange',
    'Oceania': 'yellow'
}

continent_map = {
    'Brazil': 'South America',
    'China': 'Asia',
    'United States': 'North America',
    'Canada': 'North America',
    'Japan': 'Asia',
    'United Kingdom': 'Europe',
    'Germany': 'Europe',
    'France': 'Europe',
    'Italy': 'Europe',
    'Spain': 'Europe',
    'India': 'Asia',
    'Australia': 'Oceania',
    'Russian Federation': 'Europe',
    'South Korea': 'Asia',
    'Iran': 'Asia'
}

final_df['Continent'] = final_df.index.map(continent_map)
final_df['Color'] = final_df['Continent'].map(continent_colors)

plt.figure(figsize=(14, 8))
plt.scatter(final_df['Rank'], final_df['% Renewable'], 
            s=final_df[2015] / 1e10, 
            c=final_df['Color'], alpha=0.6, edgecolors="w", linewidth=0.5)

for country in final_df.index:
    plt.text(final_df.loc[country, 'Rank'], 
             final_df.loc[country, '% Renewable'], 
             country, ha='center', va='center')

plt.xlabel('Rank')
plt.ylabel('% Renewable')
plt.title('Bubble chart')
plt.show()
```
