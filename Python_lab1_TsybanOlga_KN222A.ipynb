# Norway (country)
## General information
<span style="color:green">**Norway**</span> is a country of northern Europe that occupies the western half of the Scandinavian peninsula. Nearly half of the inhabitants of the country live in the far south, in the region around <span style="color:blue"> *Oslo*</span>, the capital.


- **Area**: 385,207 km²
- **Population**: 5.5 million (2024 estimated)
- **Government**: Unitary parliamentary constitutional monarchy

## Famous people
1. **Jo Nesbo** - Jo Nesbø is a bestselling Norwegian author known for his thrilling crime novels. He’s best known for his Harry Hole series, which has captivated readers worldwide with gripping stories of crime and investigation in Oslo.
2. **Edvard Munch** - Edvard Munch is most famous for his iconic masterpiece, ‘The Scream’. Munch was born in 1863 in a town called Adalsbruk in Eastern Norway, around 120 kilometers from Oslo.
3. **Henrik Ibsen** - Henrik Ibsen was a famous Norwegian playwright and theater director. Born in 1828 in Skein, just south of Oslo, Ibsen is regarded as one of the founders of modernism in theater, often being referred to as the ‘father of realism’.

# <span style="color:red">*Task 2 (function)*</span>

```python
import math

# function checks is an input a number
def value_check_number(num):
    if not isinstance(num, int):
        return False
    return True

# function checks is an input positive
def value_check_negative(num):
    if num < 0:
        return False
    return True


def task_2(num):
    found = False
    
    if not value_check_number(num) or not value_check_negative(num):
        return f"Please check your input {num}: it should be a number and more than 0"
    
    for i in range(0, int(math.sqrt(num)) + 1):
        a_squared = i * i
        b_squared = num - a_squared
        if b_squared >= 0:
            b = int(math.sqrt(b_squared))
            if b * b == b_squared:
                print(f'{i}^2 + {b}^2 = {num}')
                found = True
                break
                
    if not found:
        print('No such pair of natural numbers a and b exists.')
```

> ### Testing
